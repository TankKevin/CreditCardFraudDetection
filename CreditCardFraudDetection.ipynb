# 异常值处理
- 仅仅对上述相关性较大的参数进行异常值处理
- 若用全部正反例，无法得到较好分布，因此仅用正例（担心正例的异常值对模型影响会更大）
- 如果时间充分，可以考虑分别去掉正反例的异常值


````python
from scipy.stats import norm

v_positives = 'V2/V4/V11'.split('/')
v_negatives = 'V14/V12/V10/V16'.split('/')

# 正向组
fig, ax = plt.subplots(1, len(v_positives), figsize=(20, 2 * len(v_positives)))
for i, v in enumerate(v_positives):
    v_positive_dist = X_train_resampled[v].loc[y_train_resampled == 1].values
    sns.histplot(v_positive_dist, ax=ax[i], kde=True, bins=15)

plt.show()
````

````python
# 负向组
fig, ax = plt.subplots(1, len(v_negatives), figsize=(20, 2 * len(v_negatives)))
for i, v in enumerate(v_negatives):
    v_negative_dist = X_train_resampled[v].loc[y_train_resampled == 1].values
    sns.histplot(v_negative_dist, ax=ax[i], kde=True, bins=15)

plt.show()
````

````python
import numpy as np

# 去除异常值
i_to_drop = set()
for v in v_positives + v_negatives:
    v_dist = X_train_resampled[v].loc[y_train_resampled == 1]
    q25, q75 = np.percentile(v_dist, 25), np.percentile(v_dist, 75)
    iqr = q75 - q25
    cut_off = iqr * 1.5
    lower, upper = q25 - cut_off, q75 + cut_off
    v_drop = v_dist[(v_dist > upper) | (v_dist < lower)]
    i_to_drop.update(v_drop.index.values)
    outliers = [x for x in v_drop]
    print(f'Feature {v} Outliers for Fraud Cases: {len(outliers)}')
    print(f'{v} Outliers: {outliers}')

X_train_resampled = X_train_resampled.drop(i_to_drop, axis=0)
y_train_resampled = y_train_resampled.drop(i_to_drop, axis=0)
````

# 绘制学习曲线
- 简单测试训练样本是否足够达成收敛，如果达成收敛，那么可以继续 CV，否则需要增加训练样本了
- 忽略可能的报错，因为有可能训练集和验证集的样本数量不够，导致未能收敛
- 效果还不错，基本上当前的测试集是可以达成收敛的（时间关系，只测了 LR）

````python
from sklearn.model_selection import learning_curve

def plot_learning_curve(models):
    fig, axes = plt.subplots(len(models), 1, figsize=(6 * len(models), 10))

    for i, model in enumerate(models):
        print(model)
        
        # 绘制学习曲线
        ax = axes[i]
        ax.set_title('Learning Curve')
        ax.set_xlabel('Training Size')
        ax.set_ylabel('Score')
        ax.set_ylim(0.0, 1.1)
        ax.grid(True)
        
        # 计算学习曲线
        train_sizes, train_scores, test_scores = learning_curve(
            model, 
            X_train_resampled, 
            y_train_resampled, 
            cv=5, 
            train_sizes=np.linspace(0.1, 1.0, 10),
            n_jobs=-1
        )

        # 计算训练和验证集得分的均值和标准差
        train_scores_mean = np.mean(train_scores, axis=1)
        test_scores_mean = np.mean(test_scores, axis=1)

        # 绘制得分曲线
        ax.plot(train_sizes, train_scores_mean, 'o-', color='r', label='Training score')
        ax.plot(train_sizes, test_scores_mean, 'o-', color='g', label='Cross-validation score')

        ax.legend(loc='best')
    
    return fig
````

````python
fig = plot_learning_curve(classifiers)
plt.show()
````

# 讨论
## AUCPR 的计算方法
用 auc 函数计算的 AUCPR 是 0.68：
```
precision, recall, _ = precision_recall_curve(y_test, y_pred_proba)
auc_pr = auc(recall, precision)
```
使用 average_precision_score 计算的 AUCPR 是 0.66：
```
average_precision_score(y_test, y_pred_proba)
```
什么原因

### ChatGPT 回答
1. 积分方法不同：
   - auc 函数使用了数值积分（如梯形法则）来计算精确率-召回率曲线（PR曲线）下面的面积。这种方法是通过将曲线下的区域分割成小梯形并累加它们的面积来得到总面积。
   - average_precision_score 计算方法：average_precision_score 是基于每个阈值下的精确率和召回率计算的，它更接近于逐点计算PR曲线下面的面积。具体来说，它在计算每一个正样本的贡献时，考虑了该正样本作为被预测为正样本的最前位置（即阈值）时的精确率。
2. 插值方式不同：auc 函数的插值方法是线性的，而 average_precision_score 的计算方法更接近于PR曲线的插值，即它考虑了不同阈值下的精确率和召回率，可能会在某些数据点上有所不同。
3. 这些细微的计算和插值方法的不同会导致 auc 和 average_precision_score 计算出的 AUCPR 结果有所差异。一般来说，average_precision_score 被认为是更精确和可靠的PR曲线下面积的度量方法。

## AUC（ROC 曲线下的面积）和 AUCPR 的选择

在处理样本极度不平衡的问题时，选择评估指标非常重要。AUC（ROC曲线下的面积）和AUCPR（PR曲线下的面积）是常用的两种评估指标，但它们在不同场景下适用性不同。
AUC（ROC曲线下的面积）
* 优点：AUC是通过评估模型的整体性能来衡量的，它考虑了所有阈值下的表现。
* 缺点：在极度不平衡的数据集中，AUC可能会产生误导。因为在这种情况下，负样本的数量远远超过正样本，AUC可能会过分关注模型如何处理负样本，而忽视了正样本的表现。
AUCPR（PR曲线下的面积）
* 优点：AUCPR更关注正样本的表现，特别在极度不平衡的数据集中，它能够更好地反映模型对少数类的预测性能。
* 缺点：PR曲线更容易受到数据中正负样本比例的影响，因此在不同的数据集之间比较时需要小心。
### 结论
对于极度不平衡的数据集，AUCPR通常是更好的选择，因为它更关注正样本的预测效果，可以更准确地反映模型在实际应用中的性能表现。在实际使用中，建议在报告模型性能时同时使用AUC和AUCPR，以全面评估模型的表现。
